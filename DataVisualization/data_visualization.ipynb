### Problem 1

The file `anscombe.npy` contains Anscombe’s quartet.
For each section of the quartet,
- Plot the data as a scatter plot on the box [0, 20] ×[0, 13].
- Use `scipy.stats.linregress()` to calculate the slope and intercept of the least squares regression line for the data and its correlation coefficient (the first three return values).
- Plot the least squares regression line over the scatter plot on the domain $x\in[0,20]$.
- Report the mean and variance in $x$ and $y$, the slope and intercept of the regression line, and the correlation coefficient. Compare these statistics to those of the other sections.
- Describe how the section is similar to the others and how it is different.

```python
A = np.load("anscombe.npy")
m, _ = A.shape

#get data for section 1
x_1 = A[:, 0]
y_1 = A[:, 1]
#get slope, intercept, rvalue, make function
slope_1, intercept_1, rvalue_1, _, _ = linregress(x_1, y_1)
fit_1 = lambda x: slope_1 * x + intercept_1
#compute mean in x and y
mean_x_1 = (1/m) * sum(x_1)
mean_y_1 = (1/m) * sum(y_1)
#compute variance in x and y
var_x_1 = (1/(m - 1)) * sum((x_1 - mean_x_1)**2)
var_y_1 = (1/(m - 1)) * sum((y_1 - mean_y_1)**2)

#get data for section 2
x_2 = A[:, 2]
y_2 = A[:, 3]
#get slope, intercept, rvalue, make function
slope_2, intercept_2, rvalue_2, _, _ = linregress(x_2, y_2)
fit_2 = lambda x: slope_2 * x + intercept_2
#compute mean in x and y
mean_x_2 = (1/m) * sum(x_2)
mean_y_2 = (1/m) * sum(y_2)
#compute variance in x and y
var_x_2 = (1/(m - 1)) * sum((x_2 - mean_x_2)**2)
var_y_2 = (1/(m - 1)) * sum((y_2 - mean_y_2)**2)

#get data for section 3
x_3 = A[:, 4]
y_3 = A[:, 5]
#get slope, intercept, rvalue, make function
slope_3, intercept_3, rvalue_3, _, _ = linregress(x_3, y_3)
fit_3 = lambda x: slope_3 * x + intercept_3
#compute mean in x and y
mean_x_3 = (1/m) * sum(x_3)
mean_y_3 = (1/m) * sum(y_3)
#compute variance in x and y
var_x_3 = (1/(m - 1)) * sum((x_3 - mean_x_3)**2)
var_y_3 = (1/(m - 1)) * sum((y_3 - mean_y_3)**2)

#get data for section 4
x_4 = A[:, 6]
y_4 = A[:, 7]
#get slope, intercept, rvalue, make function
slope_4, intercept_4, rvalue_4, _, _ = linregress(x_4, y_4)
fit_4 = lambda x: slope_4 * x + intercept_4
#compute mean in x and y
mean_x_4 = (1/m) * sum(x_4)
mean_y_4 = (1/m) * sum(y_4)
#compute variance in x and y
var_x_4 = (1/(m - 1)) * sum((x_4 - mean_x_4)**2)
var_y_4 = (1/(m - 1)) * sum((y_4 - mean_y_4)**2)

print('Section:', '1', '2', '3', '4', sep='\t\t' )
print()
print('Mean(x):', mean_x_1, mean_x_2, mean_x_3, mean_x_4, sep='\t\t' )
print('Mean(y):', round(mean_y_1,2), round(mean_y_2,2), round(mean_y_3,2), round(mean_y_4,2), sep = '\t\t')
print('Var(x): ', round(var_x_1,3), round(var_x_2,3), round(var_x_3,3), round(var_x_4,3), sep='\t\t')
print('Slope:  ', round(slope_1, 3), round(slope_2, 3), round(slope_3, 3), round(slope_4, 3), sep='\t\t')
print('Intercept:', round(intercept_1, 3), round(intercept_2, 3), round(intercept_3, 3), round(intercept_4, 3), sep='\t\t')
print('RVal:   ', round(rvalue_1, 3), round(rvalue_2, 3), round(rvalue_3, 3), round(rvalue_4, 3), sep='\t\t')


#domain to plot the best fit functions
domain = np.linspace(0, 20, 10)

#Plot functions/data section 1
ax1 = plt.subplot(221)
ax1.plot(x_1, y_1, 'k *', markersize = 2, label='Data')
ax1.plot(domain, fit_1(domain), 'r-', label = 'Best Fit')
ax1.set_title('Section 1')
plt.axis([0, 20, 0, 13])
plt.legend(loc='best')

#Plot functions/data section 2
ax2 = plt.subplot(222)
ax2.plot(x_2, y_2, 'k *', markersize = 2, label='Data')
ax2.plot(domain, fit_2(domain), 'r-', label = 'Best Fit')
ax2.set_title('Section 2')
plt.axis([0, 20, 0, 13])
plt.legend(loc='best')

#Plot functions/data section 3
ax3 = plt.subplot(223)
ax3.plot(x_2, y_2, 'k *', markersize = 2, label='Data')
ax3.plot(domain, fit_3(domain), 'r-', label = 'Best Fit')
ax3.set_title('Section 3', y=-0.4)
plt.axis([0, 20, 0, 13])
plt.legend(loc='best')

#Plot functions/data section 4
ax4 = plt.subplot(224)
ax4.plot(x_4, y_4, 'k *', markersize = 2, label='Data')
ax4.plot(domain, fit_4(domain), 'r-', label = 'Best Fit')
ax4.set_title('Section 4', y=-0.4)
plt.axis([0, 20, 0, 13])
plt.legend(loc='best')

plt.show()
```

The $n + 1$ Bernstein basis polynomials of degree $n$ are defined as follows.

$$b_{n,v}(x) = \binom{n}{v}x^v(1-x)^{n-v},\quad v=0,\ 1,\ \ldots,\ n$$

Plot the first $10$ Bernstein basis polynomials ($n = 0,\ 1,\ 2,\ 3$) as small multiples on the domain $[0, 1] \times [0, 1]$.
Label the subplots for clarity, adjust tick marks and labels for simplicity, and set the window limits of
each plot to be the same.
Consider arranging the subplots so that the rows correspond with $n$ and the columns with $v$.

(Hint: The constant $\binom{n}{v} = \frac{n!}{v!(n-v)!}$ is called the binomial coefficient and can be efficiently computed with `scipy.special.binom()` or `scipy.misc.comb()`.)

```python
#helper function to compute bernstein polynomial
def bernstein(n, v):
    return lambda x: binom(n, v) * x**v * (1 - x)**(n-v)

#get the bernstein polynomials
poly = [bernstein(n, i) for n in range(4) for i in range(n+1)]
#plot the functions
x = np.linspace(0, 1, 100)


fig, axs = plt.subplots(4, 4, constrained_layout=True)

for i in range(4):
    for k in range(4):
        if i == 0 and k == 0:
            ax = axs[i, k]
            ax.plot(x, poly[i+k](x))
            ax.set_title('(n,v) = ' + str((i, k)))
        elif i == 0:
            break
        elif i == 1 and k <= 1:
            ax = axs[i, k]
            ax.plot(x, poly[i+k](x))
            ax.set_title('(n,v) = ' + str((i, k)))
        elif i == 1:
            break
        elif i == 2 and k <= 2:
            ax = axs[i, k]
            ax.plot(x, poly[i+k](x))
            ax.set_title('(n,v) = ' + str((i, k)))
        elif i == 2:
            break
        else:
            ax = axs[i, k]
            ax.plot(x, poly[i+k](x))
            ax.set_title('(n,v) = ' + str((i, k)))
plt.show()


```

### Problem 3

The file `MLB.npy` contains measurements from over 1,000 recent Major League Baseball players, compiled by UCLA.
Each row in the array represents a different player; the columns are the player’s height (in inches), weight (in pounds), and age (in years), in that order.

Create several visualizations to show the correlations between height, weight, and age in the MLB data set.
Use at least one scatter plot.
Adjust the marker size, plot a regression line, change the window limits, and use small multiples where appropriate.

```python
height, weight, age = np.load("MLB.npy").T

fig, axs = plt.subplots(1, 3, constrained_layout=True)

for j in range(3):
    ax = axs[j]
    if j == 0:
        ax.plot(height, weight, 'r*', markersize=1, label='data')
        ax.set_title('Height vs Weight')
        dom = np.linspace(min(height), max(height), 10)
        slope, intercept, _, _, _ = linregress(height, weight)
        regress = lambda x: slope * x + intercept
        ax.plot(dom, regress(dom), 'g--', linewidth=1, label='regression')
        ax.legend(loc='best')
        #plt.legend(loc='best')

                
    elif j == 1:
        ax.plot(height, age, 'k*', markersize=1, label='data')
        ax.set_title('Height vs Age')
        slope, intercept, _, _, _ = linregress(height, age)
        regress = lambda x: slope * x + intercept
        ax.plot(dom, regress(dom), 'b--', linewidth=1, label='regression')
        ax.legend(loc='best')
        
    else:
        ax.plot(age, weight, 'c*', markersize=1, label='data')
        ax.set_title('Age vs Weight')
        dom = np.linspace(min(age), max(age), 10)
        slope, intercept, _, _, _ = linregress(age, weight)
        regress = lambda x: slope * x + intercept
        ax.plot(dom, regress(dom), 'r--', linewidth=1, label='regression')
        ax.legend(loc='best')
    
 
            
            
plt.show()
```

### Problem 4

The file `earthquakes.npy` contains data from over 17,000 earthquakes between 2000 and 2010 that were at least a 5 on the Richter scale.
Each row in the array represents a different earthquake; the columns are the earthquake’s date (as a fraction of the year), magnitude (on the Richter scale), longitude, and latitude, in that order.

Describe the data with two or three visualizations, including line plots, scatter plots, and histograms as appropriate.
Your plots should answer the following questions:
1. How many earthquakes happened every year?
2. How often do stronger earthquakes happen compared to weaker ones?
3. Where do earthquakes happen? Where do the strongest earthquakes happen?

Hint: Use `plt.axis("equal")` or `ax.set_aspect("equal")` to fix the aspect ratio, which may improve comparisons between longitude and latitude.

```python
years, magnitudes, longitude, latitude = np.load("earthquakes.npy").T

fig, axs = plt.subplots(2, 2, constrained_layout=True)
#fig, axs = plt.subplots(4, 1)
ax = axs[0, 0]
ax.hist(years, bins=np.arange(1999.5, 2010.5), label='Years')
ax.set_title('Earthquakes per Year')
ax = axs[0, 1]
ax.hist(magnitudes, bins=np.arange(0, 11), label='Magnitudes', density=True)
ax.set_title('Magnitudes')
#plt.plot(years, magnitudes, '.')
#print(max(years))
#plt.xlabel("Year")
#plt.ylabel("Magnitude")
ax = axs[1, 0]
ax.plot(longitude, latitude, '.', markersize=1)
ax.set_title('Locations')
ax.set_xlabel('longitude')
ax.set_ylabel('latitude')
ax.set_aspect('equal')


#getting longitude and latitude values for specific magnitude range
mag_vals = [5, 6, 7, 8, 9]
lat = []
long = []
n = len(longitude)

for j in range(len(mag_vals)):
    if j == 0:
        mag_long = np.array([longitude[i] for i in range(n) if magnitudes[i] <= 5])
        mag_lat = np.array([latitude[i] for i in range(n) if magnitudes[i] <= 5])
    else:
        mag_long = np.array([longitude[i] for i in range(n) if magnitudes[i] <= mag_vals[j] and magnitudes[i] > mag_vals[j-1]])
        mag_lat = np.array([latitude[i] for i in range(n) if magnitudes[i] <= mag_vals[j] and magnitudes[i] > mag_vals[j-1]])
    long.append(mag_long)
    lat.append(mag_lat)

colors = ['b', 'g', 'r', 'c', 'k']
    
ax = axs[1, 1]

for i in range(len(mag_vals)):
    color = colors[i] + '.'
    lab = 'Magnitude: ' + str(mag_vals[i])
    if mag_vals[i] == 9:
        lab += '+'
    if i == 0:
        s = 1
        a = 0.5
    else:
        s = 1
        a = 0.5
    ax.plot(long[i], lat[i], color, markersize=s, alpha=a, label=lab)

ax.set_title('Magnitudes per Location')
ax.set_xlabel('longitude')
ax.set_ylabel('latitude')
ax.legend(loc='best', prop={'size': 3})
plt.show()
```

### Problem 5

The _Rosenbrock function_ is defined as follows. $$f(x,y)=(1−x)^2 +100(y−x^2)^2$$
The minimum value of $f$ is $0$, which occurs at the point $(1, 1)$ at the bottom of a steep, banana-shaped valley of the function.

Use a heat map and a contour plot to visualize the Rosenbrock function.
Plot the minimizer $(1, 1)$.
Use a different sequential colormap for each visualization.

```python
rosenbrock = lambda x, y: (1 - x)**2 + 100 * (y - x**2)**2 #make a lambda function for rosenbrock function
side = np.linspace(-20, 20, 1000)#this will make one side of the square we are interested in
X, Y = np.meshgrid(side, side)
func_vals = rosenbrock(X, Y)

fig, axs = plt.subplots(1,2)

ax = axs[0]
ax.pcolormesh(X, Y, func_vals, cmap='viridis')
ax.plot(1, 1, 0, 'r.', markersize=6, label='min')
ax.set_xlim(-20, 20)
ax.set_ylim(-20, 20)
ax.set_title('Heat Map')
ax = axs[1]
ax.contour(X, Y, func_vals, [-20, -10, 0, 10, 20], cmap='inferno')
ax.plot(1, 1, 0, 'r.', markersize=3)
ax.set_xlim(-4, 4)
ax.set_ylim(-1, 5)
ax.set_title('Contour Map')
plt.show()
```
